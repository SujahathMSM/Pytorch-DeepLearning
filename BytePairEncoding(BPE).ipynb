```python
from importlib_metadata import version
import tiktoken
print(version("tiktoken"))
```

```python
tokenizer = tiktoken.get_encoding('gpt2')
```

```python
word = (
    "Hello, do you like tea <|endoftext|> In the sunlight terraces"
    "of someUnknownPlace"
)
```

```python
integers = tokenizer.encode(word, allowed_special={"<|endoftext|>"})
print(integers)
```

```python
strings = tokenizer.decode(integers)
print(strings)
```

```python
new_word = "Akwirw ier"
int_tokens = tokenizer.encode(new_word)
print(int_tokens)
```

```python
print(tokenizer.decode(int_tokens))
```
